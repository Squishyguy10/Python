```python
def predict(x, w, b):
    return x * w + b
```

```python
def loss(x, y, w, b):
    predictions = predict(x, w, b)
    return np.average(abs(predictions - y))
```

```python
def gradient(x, y, w, b):
    wgrad = np.average(2*x*(predict(x, w, b) - y))
    bgrad = np.average(2*(predict(x, w, b) - y))
    return (wgrad, bgrad)
```

```python
gradient(x, y, 2, 3)
```

```python
def train(x, y, itr, lr):
    w = b = 0
    for i in range(itr):
        print("Itr %4d => Loss: %.6f" % (i, loss(x, y, w, b)))
        wgrad, bgrad = gradient(x, y, w, b)
        w = wgrad * lr
        b = bgrad * lr
    return w, b
```

```python
w, b = train(x, y, 100000, 0.0001)
```
